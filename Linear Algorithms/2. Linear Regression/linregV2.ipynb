```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

in linear regression, the target variable is assumed to follow a linear function of one or more predictor variables,
, plus some random error.

$${Y}_{n} = {\beta}_{0} + {\beta}_{1}X_{n1} + ... + {\beta}_{e}X_{ne} +{E}$$

Where ${Y}_{n}$ is the target variable ${\beta}_{0}  $ is the Intercept ${X}_{n}$ is the observation and ${\beta}_{1 ... n}$ are the coefficients .

$$ {X}_{n} = {(} {\;}  {1} {\;} {X}_{n1} {\;} {\dots} {\;} {X}_{nD} {\;} {)}^{T}$$
$$ {\beta} = {(} {\;} {\beta}_{0} {\;} {\beta}_{1} {\;} ... {\;} {\beta}_{D} {\;} {)}^{T}$$

1 is the intercept, thus we can write as the following

$$ {Y}_{n} = {\;\beta}^{T}{X}_{n\;} + {\;\in}  $$

```python
# Lets Try to make it

#Generate the data 
np.random.seed(5)
N = 20
beta0 = -2
beta1 = 4
x = np.random.randn(N) # Observations
e = np.random.randn(N) # Error
y = beta0 + beta1*x + e # Training

# Testing the model
true_x = np.linspace(min(x), max(x), 100)
true_y = beta0 + beta1 * true_x

## Plotting

f,ax = plt.subplots(1,2,figsize=(10,6))
ax[0].scatter(x,y)
ax[0].set(xlabel="Values", ylabel="y_training" )
ax[0].set_title(fr"$y = ${beta0} + ${beta1} + epsilon$",fontsize=16)
ax[1].scatter(true_x,true_y)
ax[1].set(xlabel="Values", ylabel="y_traing")
ax[0].legend(loc=4)
```
